```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x = np.random.normal(size=(2 * n, 2))
x[:n, 0] += 2
x[:n, 1] -= 2
y = np.concatenate((np.repeat(1, n), np.repeat(2, n)))
x_train = pd.DataFrame({'x1': x[:, 0], 'x2': x[:, 1], 'y': y})

plt.scatter(x_train[y == 1]['x1'], x_train[y == 1]['x2'],c='red',label='Class 1')
plt.scatter(x_train[y == 2]['x1'], x_train[y == 2]['x2'],c='blue',label='Class 2')

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

# criterion default='gini'
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train.drop('y',axis=1),x_train['y'])
```

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(x_train.drop('y', axis=1))
acc=accuracy_score(x_train['y'],y_pred)
acc
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10,10))
plot_tree(dt)
plt.show()
```

```python
dt=DecisionTreeClassifier(max_depth=3,criterion='entropy')
dt.fit(x_train.drop('y',axis=1),x_train['y'])

y_pred = dt.predict(x_train.drop('y', axis=1))
acc=accuracy_score(x_train['y'],y_pred)
acc
```

```python
plt.figure(figsize=(10,10))
plot_tree(dt)
plt.show()
```
